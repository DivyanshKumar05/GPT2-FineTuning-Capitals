```python
from google.colab import drive
import torch
from transformers import GPT2LMHeadModel, GPT2Tokenizer

# 1. Mount your Google Drive so Colab can see the files
drive.mount('/content/drive')

# 2. Set up your hardware accelerator
device = "cuda" if torch.cuda.is_available() else "cpu"

# 3. Load the model and tokenizer straight from your Drive folder
model_path = '/content/drive/MyDrive/gpt2_finetuned_capitals'

tokenizer = GPT2Tokenizer.from_pretrained(model_path)
model = GPT2LMHeadModel.from_pretrained(model_path).to(device)

print("🚀 Success! Your fine-tuned model is loaded and ready.")
```

```python
# 4. Test the model to showcase its performance
test_prompt = "Question: What is the capital of Japan?\nAnswer:"

# Tokenize the input prompt
inputs = tokenizer(test_prompt, return_tensors="pt").to(device)

# Generate the answer using Greedy Decoding (factual, no randomness)
outputs = model.generate(
    inputs['input_ids'],
    attention_mask=inputs['attention_mask'],
    max_new_tokens=15,
    do_sample=False,
    pad_token_id=tokenizer.eos_token_id
)

# Decode and print the result
result = tokenizer.decode(outputs[0], skip_special_tokens=True)
print("--- SHOWCASE MODEL OUTPUT ---")
print(result)
```
